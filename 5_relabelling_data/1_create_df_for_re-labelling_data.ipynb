```python
### Read IDs of strongly misclassified points from file 
### This is generated by running the notebook 2_anamalous_points.ipynb and saving IDs for select categories
# fname='/global/u1/v/vpa/project/jpt_notebooks/Cosmology/Supernova-cnn/cnn_supernova/3_analyze_results_jpt_notebooks/Misclassified_IDs_cnn_model9.csv'
fname='/global/u1/v/vpa/project/jpt_notebooks/Cosmology/Supernova-cnn/cnn_supernova/3_analyze_results_jpt_notebooks/Misclassified_IDs_cnn_model3_fulldataset.csv'
IDs=np.loadtxt(fname).astype(int)
print(IDs[:6])
```

```python
### Use select IDs
df3=df2[df2.ID.isin(IDs)]
df3.shape
del(df2)
df3.head()
```

```python
keys=['temp','srch','diff']
imgs=np.array([[plt.imread(df3[(df3.ID==iD)&(df3.filename.str.startswith(key))]['file path'].values[0]) for key in keys] for iD in IDs])
print(imgs.shape)
```

```python
keys=['temp','srch','diff']

cols=['ID','old_label','changed_label','final_label','imgs']
df_labels=pd.DataFrame(columns=cols)
df_labels

for iD in IDs:
    old_label=df3[(df3.ID==iD)&(df3.filename.str.startswith('temp'))]['Label'].values[0]
#     print(old_label)
    ### Create image array
    imgs=np.array([plt.imread(df3[(df3.ID==iD)&(df3.filename.str.startswith(key))]['file path'].values[0]) for key in keys])
    dict1={'ID':np.int(iD),'old_label':old_label,'changed_label':(1+old_label)%2,'final_label':'','imgs':imgs}
#     print(dict1)
    df_labels=df_labels.append(dict1,ignore_index=True)
    
df_labels.head()
```

```python
def f_plot(df,iD):
    fig=plt.figure()

    for i in range(3):
        fig.add_subplot(1,3,i+1)
        plt.imshow(df.iloc[iD].imgs[i,:,:])
        plt.title(keys[i])

f_plot(df_labels,5)
```
